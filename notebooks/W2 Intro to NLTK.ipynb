This is an object of a new type, called `Text`:

````python
# Let's see what gets printed when we try printing this object
print(wall_street)
````

````python
# What about its type?
type(wall_street)
````

To access the actual text, we can use the `[]` operator just as if our text were a list. For example, say we wanted to print the first 100 "words" in our Wall Street Journal text. We could do this with:

````python
# Create a list of the first 100 "words"
# What constitutes a "word" here will be the topic of a future class
# For now, notice how punctuation markers count as separate "words"
first_words = [wall_street[i] for i in range(100)]

print(first_words)
````

Another similarity between `Text`s and `list`s is that it has a "length", which can be accessed through the `len` function:

````python
len(wall_street)
````

They can also be sorted...

````python
sorted(wall_street)
````

... and converted into a `set`...

_(A `set` is another data structure, like our `list`s and `tuple`s, that stores any type of information and does not keep any ordering between the elements, nor duplicate elements)_

````python
set(wall_street)
````

... and if we want a list of all the unique "words" we have in our text, we can transform out text into a `set` and print them sorted:

````python
sorted(set(wall_street))
````

And if we want to know how many unique "words" are there, then we can replace `sorted` by `len`:

````python
len(set(wall_street))
````

This way, we can get an idea about the "lexical diversity" of a given text by getting the number of unique words and dividing by the number of words:

````python
# Calculates the "lexical diversity" of a given text.
len(set(wall_street)) / len(wall_street)
````

Let's think a little about the calculation above.
  * If each word in the text is different from the other, the numerator will be big, and the final number will approach 1;
  * If all words are the same, the numerator will become 1, and the final number will approach 0.

This way, we can test some interesting linguistic hypotheses. For example, I would expect that books are more lexically diverse than newspapers. We could compare the lexical diversity above with that of Moby Dick:

````python
# Back in the beginning of the class, when we imported nltk.book, we got a list of all
# the texts available in it
moby_dick = nltk.book.text1

# Surprisingly, Moby Dick has a lower proportion of unique words =O
len(set(moby_dick)) / len(moby_dick)
````

Of course, this is not a great measure if we only test one book. It would be interesting to have a bigger and more diverse set of books to try out this hypothesis.

Changing topics a little, another kind of thing we might want to do is to count the number of times a given word appears in the corpus. This can be done with the `count` function of our `Text` object:

````python
wall_street.count('Brazil')
````

Finally, it is also possible to find the first occurrence of a given word with the `.index()` method. For example, if we wanted to find the first occurrence of the word `Brazil` in our `wall_street` text, we could do:

````python
wall_street.index('Brazil')
````

Then, if we wanted to print the context in which this word appeared, we could print the words surrounding this index:

````python
print(wall_street[5380:5400])
````

## Concordances

From Wikipedia:

> A concordance is an alphabetical list of the principal words used in a book or body of work, listing every instance of each word with its immediate context.
> (https://en.wikipedia.org/wiki/Concordance_%28publishing%29)


There are several ways in which you might want to present your concordance: (the examples below are from Wikipedia's [_Key Word in Context_](https://en.wikipedia.org/wiki/Key_Word_in_Context) article)

[Keyword in context (KWIC)](https://en.wikipedia.org/wiki/File:KWIC.png)

[Keyword alongside context (KWAC)](https://en.wikipedia.org/wiki/File:KWAC.png)

[Keyword out of context (KWOC)](https://en.wikipedia.org/wiki/File:KWOC.png)

In the examples above, the show topics of interest. Even with computers, producing topic-based concordances is very labor intensive, and will not be our focus here. Instead, with `nltk`, our goal will be to look for all occurences of a given word of interest. For example, we might look for all occurrences of the word `Brazil` in our `wall_street` text:

````python
# nltk always displays KWIC
wall_street.concordance("Brazil")
````

One problem with `nltk`'s concordances is that it doesn't care about the sense of the word we look for. One word in English that happens to appear with a lot of senses is the word `get`. Let's take a look at the concordance we will get for it:

````python
wall_street.concordance("get")
````

As you can see, `get` can assume the meaning of "acquire" (as in _get a taste_ or _get AIDS_), denote some sort of passive (as in _get burned_ or _get elected_), denote some sort of movement (as in _get across_ or _get around_), or even mean "become" (as in _get excited_).

#### Additional Resources on Concordances

You can find a lot of useful concordances in the internet, many of which are related to a corpus. For example:

 * A concordance from the British National Corpus: http://corpus.byu.edu/bnc/
 * For the Contemporary Corpus of American English: https://www.english-corpora.org/coca/

## Words with similar contexts

_(part of the explanation in this section is based on https://stackoverflow.com/questions/43438008/difference-between-similar-and-concordance-in-nltk)_


If we look at our concordances, we might ask ourselves: which other words appear in a similar context in our corpus?

For example, we might ask what other words appear in a context similar to that of `Brazil`. The `nltk` provides a function to do just that:

````python
wall_street.similar('Brazil')
````

Let's take a look at the concordance of `ford` (the last word among those similar to Brazil):

````python
wall_street.concordance('ford')
````

For `nltk`, "context" here means "one word before and one word after". So, for the first line above, the context of `Ford` is `of ______ Motor`.

Now... from these two concordances, it is hard to look for how many lines had the same context. To make our life easier, `nltk` has the `common_contexts` function:

````python
wall_street.common_contexts(['Brazil', 'Ford'])
````

This means that both `that Brazil a` and `that Ford a` appear in the Wall Street Corpus.

To have a notion of how distributed the words in the corpus are, it might be useful to plot a dispersion plot. We can do it with the `dispersion_plot` function. In the graph below, we also look for the dispersion of `Volkswagen`, `computer` and `Korea`:

````python
import matplotlib
````

````python
# Before running the following line, you'll need to install matplotlib.
# Open a terminal, activate your virtualenv, and then call
#     pip install matplotlib
# then run
# import matplotlib
# and only then run the line below

wall_street.dispersion_plot(['Brazil', 'Ford', 'Volkswagen', 'computer', 'Korea'])
````

## Frequency distributions

The `nltk` also provides functions to calculate how frequent all the words in a given text are. The code below calculates the frequency of each word in the Wall Street Journal text and assigns it to the variable `frequencies_wsj`:

````python
frequencies_wsj = nltk.FreqDist(wall_street)
frequencies_wsj
````

Of course, we can do the same with any other element of the type `Text`:

````python
frequencies_md = nltk.FreqDist(moby_dick)
````

We can, for example, get the 50 most common words in our `wall_street` object:

````python
frequencies_wsj.most_common(50)
````

We can even plot these frequencies! Say, for example, we wanted to plot the frequencies of the 30 most common words:

````python
import matplotlib.pylab as plt
plt.figure(dpi=600)

# Word frequencies generally follow a Zipf's distribution
frequencies_wsj.plot(30, cumulative=False)
````

Here, `cumulative=True` would produce a graph such that values to the right always include the sum of all the element to their left. That is...

````python
frequencies_wsj.plot(30, cumulative=True)
````

Again, producing this sort of graph allows us to answer interesting questions about the texts we are analysing. For example, notice how most of the words in the graph above (i.e., most of the "most common words" in our `wall_street` object) are function words, such as prepositions or articles.

The `FreqDist` object has several methods about which we are not going to talk much here. They should mostly work in an intuitive way:

| Example |	Description |
|-|-|
| fdist = FreqDist(samples) |	create a frequency distribution containing the given samples |
| fdist[sample] += 1 |	increment the count for this sample |
| fdist['monstrous'] |	count of the number of times a given sample occurred |
| fdist.freq('monstrous') |	frequency of a given sample |
| fdist.N() |	total number of samples |
| fdist.most_common(n) |	the n most common samples and their frequencies |
| for sample in fdist:	| iterate over the samples |
| fdist.max()	| sample with the greatest count |
| fdist.tabulate()	| tabulate the frequency distribution |
| fdist.plot()	| graphical plot of the frequency distribution |
| fdist.plot(cumulative=True)	| cumulative plot of the frequency distribution |
| fdist1 = fdist2	| update fdist1 with counts from fdist2 (*be careful!!*) |
| fdist1 < fdist2	| test if samples in fdist1 occur less frequently than in fdist2 |

One interesting method, though, is the `hapaxes` method.

A [_hapax legomenon_](https://en.wikipedia.org/wiki/Hapax_legomenon) (plural _hapax legomena_) is a word that appears only once in a body of text. These are words that we might want to give special treatment in our analysis. Since they appear only once, it is not easy to find their meaning, nor in which contexts they commonly appear. We might even want to exclude them from analysis, or check if they are not just input errors.

Let's look for hapax legomena in our `wall_street` text:

````python
frequencies_wsj.hapaxes()
````

Before finishing this topic, we will talk about one last problem: that of finding _Collocations_.

Collocation is a term often used meaning different things. The [Wikipedia](https://en.wikipedia.org/wiki/Collocation) defines them as _a sequence of words or terms that co-occur more often than would be expected by chance_, giving examples such as _strong tea_ and _powerful computer_, which would sound strange if referred to as _powerful tea_ or _strong computer_. [Other authors](https://nlp.stanford.edu/fsnlp/promo/colloc.pdf), on the other hand, require collocations to satisfy other other requirements, like non-compositionality (i.e., to have a meaning that can not be derived from the parts) or non-modifiability (e.g., it wouldn't be possible to generate, by analogy with _kick the bucket_, meaning die, something like _strike the vessel_, also meaning die).

Here, we only care about the computational problem of finding any group of two or more words that appear often together, leaving the linguistic problem aside.

To find these pairs of words, `Text` objects provide a quite useful method:

````python
wall_street.collocations()
````
